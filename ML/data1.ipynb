```python
import numpy as np
import pandas as pd

# Load the galaxy catalogue data
data = np.load('galaxy_catalogue.npy', allow_pickle=True)

# Convert the structured array to a DataFrame
df = pd.DataFrame({
    'u-g': [row['u-g'] for row in data],
    'g-r': [row['g-r'] for row in data],
    'r-i': [row['r-i'] for row in data],
    'i-z': [row['i-z'] for row in data],
    'eccentricity': [row['ecc'] for row in data],
    'm4_u': [row['m4_u'] for row in data],
    'm4_g': [row['m4_g'] for row in data],
    'm4_r': [row['m4_r'] for row in data],
    'm4_i': [row['m4_i'] for row in data],
    'm4_z': [row['m4_z'] for row in data],
    'petroR50_u': [row['petroR50_u'] for row in data],
    'petroR50_r': [row['petroR50_r'] for row in data],
    'petroR50_z': [row['petroR50_z'] for row in data],
    'petroR90_u': [row['petroR90_u'] for row in data],
    'petroR90_r': [row['petroR90_r'] for row in data],
    'petroR90_z': [row['petroR90_z'] for row in data],
    'class': [row['class'] for row in data]  # Target label
})

# Display the DataFrame
print(df.head())
```

```python
df['ratio_u_g_r_i'] = df['u-g'] / df['r-i']
df['ratio_u_g_g_r'] = df['u-g'] / df['g-r']
df['ratio_g_r_r_i'] = df['g-r'] / df['r-i']
df['ratio_r_i_i_z'] = df['r-i'] / df['i-z']
```

```python
df['median_magnitude'] = df[['m4_u', 'm4_g', 'm4_r', 'm4_i', 'm4_z']].median(axis=1)
df['mean_eccentricity'] = df[['m4_u', 'm4_g', 'm4_r', 'm4_i', 'm4_z']].mean(axis=1)
df['eccentricity_variation'] = df[['m4_u', 'm4_g', 'm4_r', 'm4_i', 'm4_z']].std(axis=1)
df['eccentricity_range'] = df[['m4_u', 'm4_g', 'm4_r', 'm4_i', 'm4_z']].max(axis=1) - df[['m4_u', 'm4_g', 'm4_r', 'm4_i', 'm4_z']].min(axis=1)
df['mean_normalized'] = df[['m4_u', 'm4_g', 'm4_r', 'm4_i', 'm4_z']].apply(
    lambda x: (x - x.min()) / (x.max() - x.min())
).mean(axis=1)
```

```python
df['concentration_u'] = df['petroR50_u'] / df['petroR90_u']
df['concentration_r'] = df['petroR50_r'] / df['petroR90_r']
df['concentration_z'] = df['petroR50_z'] / df['petroR90_z']
```

```python
import optuna
import xgboost as xgb


def objective(trial):
    # Suggest hyperparameters
    params = {
        'objective': 'multi:softmax',
        'num_class': len(set(y)),  # Number of classes
        'learning_rate': trial.suggest_float('learning_rate', 1e-5, 1e-1, log=True),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'min_child_weight': trial.suggest_int('min_child_weight', 1, 10),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'gamma': trial.suggest_float('gamma', 0, 5),
        'reg_alpha': trial.suggest_float('reg_alpha', 1e-5, 1e-1, log=True),
        'reg_lambda': trial.suggest_float('reg_lambda', 1e-5, 1e-1, log=True),
        'scale_pos_weight': trial.suggest_float('scale_pos_weight', 1, 10),  # Adjust if needed
        'random_state': 42,
        'use_label_encoder': False
    }
    
    # Create the model
    model = xgb.XGBClassifier(**params)

    # Fit the model
    model.fit(X_train, y_train)

    # Make predictions
    y_pred = model.predict(X_test)

    # Calculate accuracy
    accuracy = accuracy_score(y_test, y_pred)
    return accuracy
```

```python
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=300)
```

## **Study of HyperParams**

```python
print("Best hyperparameters: ", study.best_params)
print("Best accuracy: ", study.best_value)
```

```python
import xgboost as xgb
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Best hyperparameters obtained from Optuna
best_params =  {'learning_rate': 0.010441680065432429,
                'max_depth': 7,
                'min_child_weight': 1,
                'subsample': 0.6382732741365813,
                'colsample_bytree': 0.841791294785763,
                'gamma': 0.9078212651627887,
                'reg_alpha': 0.009260871302743046, 
                'reg_lambda': 0.009404349567886858, 
                'scale_pos_weight': 8.163820500567693
                }

# Step 1: Initialize the XGBoost model with best parameters
model = xgb.XGBClassifier(**best_params)

# Step 2: Train the model
model.fit(X_train, y_train)

# Step 3: Make predictions
y_pred_test = model.predict(X_test)
y_pred_train = model.predict(X_train)

# Step 4: Evaluate the model
accuracy_test = accuracy_score(y_test, y_pred_test)
accuracy_train = accuracy_score(y_train, y_pred_train)

print("Test Accuracy:", accuracy_test)
print("Train Accuracy:", accuracy_train)

# Step 5: Display additional metrics
print("Classification Report (Test):")
print(classification_report(y_test, y_pred_test))

print("Confusion Matrix (Test):")
print(confusion_matrix(y_test, y_pred_test))
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc, confusion_matrix, cohen_kappa_score

# Assuming you have your true labels and predicted probabilities
# y_test = ...
# y_pred_proba = ... # Predicted probabilities from the model

# For multi-class, you may need to use OneVsRestClassifier or OneVsAll approach
# For example, y_pred_proba could be a 2D array with shape (n_samples, n_classes)
y_pred_proba = model.predict_proba(X_test)
# Calculate ROC AUC for each class
n_classes = 3  # Update based on your number of classes
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    fpr[i], tpr[i], _ = roc_curve(y_test == i, y_pred_proba[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plotting the ROC Curve
plt.figure(figsize=(8, 6))
for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], label='ROC curve (class {0}) (AUC = {1:0.2f})'.format(i, roc_auc[i]))

plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc='lower right')
plt.grid()
plt.show()

# Calculate Kappa Statistic
y_pred = np.argmax(y_pred_proba, axis=1)  # Get the predicted class
kappa_score = cohen_kappa_score(y_test, y_pred)

print(f"Kappa Statistic: {kappa_score:.4f}")
```
